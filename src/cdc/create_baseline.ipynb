```python
from pyspark.sql import SparkSession

tables = ["clientes", "produtos", "transacoes", "transacao_produto"]

for t in tables:
    src = f"/Volumes/raw/sistema_pontos/full_load/{t}"
    dst = f"/Volumes/raw/sistema_pontos/baseline/{t}"

    df = spark.read.parquet(src)
    df.coalesce(1).write.mode("overwrite").parquet(dst)

    print(f"✅ baseline criado: {t}")
## Depois nao é mais necessario criar uma nova baseline pois ela ja é atualizado pelo cdc_ingestion
```
